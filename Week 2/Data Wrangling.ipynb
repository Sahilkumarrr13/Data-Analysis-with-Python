```python
import pandas as pd 
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(path, header = None)
```

```python
print("The first 5 rows of table are ")
df.head(5)
```

```python
print("The bottom 10 rows are")
df.tail(10)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("Headers \n", headers)
```

```python
df.columns = headers
df.head(10)
```

```python
df.info()
```

## Data Wrangling

```python
import numpy as np

df.replace("?",np.nan, inplace=True)

df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
df
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_nom_loss = df["normalized-losses"].astype('float').mean(axis=0)
df["normalized-losses"].replace(np.nan, avg_nom_loss, inplace=True)
```

```python
avg_stroke = df["stroke"].astype('float').mean(axis = 0)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
avg_bore = df["bore"].astype('float').mean(axis = 0)
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_horsepower = df["horsepower"].astype('float').mean(axis = 0)
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm = df["peak-rpm"].astype('float').mean(axis = 0)
df["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df["num-of-doors"].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
df.dropna(subset=["price"], axis = 0, inplace=True)
```

```python
df.reset_index()
```

```python
new = df.isnull()
for columns in new.columns.values.tolist():
    print(column)
    print(new[column].value_counts())
    print("")
```

> We can see that the data is cleaned with no missing values

### The next step is changing the data types to correct ones

```python
df.dtypes
```

```python
df.head(2)
```

```python
df["normalized-losses"] = df["normalized-losses"].astype('int')
df[["bore","stroke"]] = df[["bore","stroke"]].astype('float')
df[["horsepower","peak-rpm"]] = df[["horsepower","peak-rpm"]].astype('int64')
```

```python
df.dtypes
```

# Data Standardization

```python
df.head()
```

```python
df["city-mpg"] = 235/df["city-mpg"]
```

```python
df.head()
```

```python
df.rename(columns={"city-mpg":"city-L/100km"}, inplace=True)
```

```python
df["highway-L/100km"] = 235/df["highway-mpg"]
```

```python
df
```

## Data Normalisation

>Normalising length, height, width

```python
df["height"]
```

```python
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()
```

```python
df["height"]
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower")
```

## Binning

```python
bins = np.linspace(df["horsepower"].min(), df["horsepower"].max(), 4)
bins
```

```python
group_names = ["Low", "Medium", "High"]
```

```python
df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels=group_names, include_lowest=True)
df[["horsepower","horsepower-binned"]].head(5)
```

```python
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower-binned"])

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower-binned")
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.bar(group_names, df["horsepower-binned"].value_counts())

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower-binned")
```

## Data Indicator Variable

```python
df_dummy_1 = pd.get_dummies(df["fuel-type"])
df_dummy_1
```

```python
df_dummy_1.rename(columns={'gas':'fuel-type-gas','diesel':'fuel-type-diesel'}, inplace=True)
df_dummy_1
```

Combine the dummy with the main dataframe, df


```python
df = pd.concat([df,df_dummy_1],axis=1)
df.drop("fuel-type", axis=1, inplace = True)
```

```python
df.rename(columns={'std':'aspiration-std','turbo':'aspiration-turbo'}, inplace=True)
df = pd.concat([df,df_dummy_2],axis=1)
```

```python
df.drop("aspiration", axis=1, inplace=True)
```
